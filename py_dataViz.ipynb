```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x  = np.linspace(0, 10, 25)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, '-ok', color = 'b')
plt.show()
```

```python
fig, ax = plt.subplots(2)
x1 = np.random.randn(100)
y1 = np.random.randn(100)
ax[0].plot(x,y, '--', label  = "sinus")
ax[1].plot(x,np.cos(x), '-', label = "cosine")
ax[0].set(xlabel = "x", ylabel = "y", title = "sinus")
ax[0].set(xlabel = "x", ylabel = "y", title = "cosine")
plt.show()
```

```python
plt.plot(x,y, '--', label  = "sinus")
plt.plot(x,np.cos(x), '-', label = "cosine")
plt.legend()
```

```python
x = np.linspace(0,10,40)
y = np.sin(x) + 0.8 * np.random.randn(40)
plt.errorbar(x, y, yerr=0.8, fmt='o', capsize=5)
```
